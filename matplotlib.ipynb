```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([0,1,2,3,4],[0,2,4,6,8],label='2x',c="red",linewidth = 3,marker='*',markersize= 10 , markeredgecolor='blue',linestyle='--')
  #           x      y
plt.title("Sample plot")
plt.xlabel('x label', fontdict = {'fontname':'serif','fontsize':20})  # Add an x-label to the axes.
plt.ylabel('y label')
plt.xticks([0,1,2,3,4])# to change what numbers  show on x axis
#plt.yticks([0,2,4,6,8,10])#same for y
plt.legend()
x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:5],x2[:5]**2,'c', label='x^2')
plt.plot(x2[4:],x2[4:]**2,'r--')
plt.legend()
plt.figure(figsize=(5,3),dpi=5000)#to change size
plt.savefig('x.png',dpi=3000)

plt.show() #<------- use this to remove this line : [<matplotlib.lines.Line2D at 0x7bfc234586d0>]
```

```python
labels=['A','B','C']
values = [1,4,2]
bars=plt.bar(labels,values)
plt.figure(figsize=(6,4))
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')
plt.show()
patterns=['/','O','*']
for bar in bars :
  bar.set_hatch(patterns.pop(0))
```

```python
fifa= pd.read_csv('fifa_data.csv')
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins,color="red")
plt.xticks(bins)
plt.xlabel("nb of players")
plt.ylabel("skill lvl")
plt.title("distribution of player skills in fifa")
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels = ['Left','Right']
colors=['green','blue']
plt.title('foot preference of fifa players')
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f')
plt.show()
```

```python
fifa.Weight = [int (x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

   # initial weight was string since it had lbs so we need to transform to an int
light =fifa.loc[(fifa.Weight < 125)].count()[0]
plt.style.use('ggplot')
medium = fifa.loc[(fifa.Weight >=125)& (fifa.Weight < 150 )].count()[0]
heavy = fifa.loc[(fifa.Weight >=150)& (fifa.Weight < 200 )].count()[0]
weights = [light,medium,heavy]
labels = ['light','medium','heavy']
explode= (.5,.1,.1)
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.title('weight in lbs ')
plt.show()
```
